```python
# Import libraries
import pandas as pd
import json
from pathlib import Path
```

```python
with open('pools_list.json') as f:
  data = json.load(f)
data_df = pd.DataFrame(data)
data_df.head()
```

```python
print(len(data_df.columns))
```

```python
column_count = [*range((len(data_df.columns)))]
print(column_count)
```

```python
pools_df = pd.DataFrame()
list_num = [1,2,3]
for i in column_count:
    pool_dict = data_df.iloc[0,[i]].to_dict()
    dict_values = pool_dict.values()
    temp_df = pd.DataFrame(dict_values)
    #print(temp_df)
    pools_df = pools_df.append(temp_df)
```

```python
pools_df.head()
```

```python
pools_df.columns
```

```python
filtered_df = pools_df[['id','db_ticker','total_stake','rewards_epoch','tax_fix','active_stake','pledge','tax_real','tax_ratio','roa','roa_lifetime']]
filtered_df.head()
```

```python
def reward_calculator(pool):
    total_volume = pool['active_stake'] + my_wallet
    my_percent = my_wallet / total_volume
    pool_rewards = (pool['roa']/100) * total_volume
    my_rewards = (pool_rewards * my_percent) * (1 - pool['tax_ratio'])
    return my_rewards
```

```python
test_df = filtered_df[:5].copy()
```

```python
test_df
```

```python
test_df['roa'] = pd.to_numeric(test_df['roa'],errors = 'coerce')
test_df['tax_ratio'] = pd.to_numeric(test_df['tax_ratio'],errors = 'coerce')
test_df.dtypes
```

```python
my_wallet = 500
test_df['my_rewards'] = test_df.apply(reward_calculator, axis=1)
```

```python
test_df.sort_values(by=['my_rewards'], ascending=False)
```
